```python
# return weights by gradient descent algorithm 

# precision denotes the minimum loss value at which we stop 
# None means it does not stop until all iterations have been
# run through. 

# batch gradient descent

import numpy as np

def gradient_descent(X,y,alpha,T,precision=1e-06):
    x0 = []

    # craeting a new column that represents x0
    for i in range(X.shape[0]):
        x0.append([1])
        
    x0=np.array(x0)
    X = np.concatenate((x0,X),axis=1)
 
    # NOTICE: STARTING weights is always 1
    
    w = np.ones(X.shape[1]).reshape(-1,1)
    
    cost = []
        
    for i in range(T):
        #print(w)
        hx = 1/(1+np.exp(-np.dot(X,w)))
        #print(hx)
        w = w.reshape(w.shape[1],w.shape[0])
        w = w + alpha*sum((y-hx)*hx*(1-hx)*X)
        w = w.reshape(w.shape[1],w.shape[0])
        
        # cost function update
        cost.append(sum((y-hx)**2))
        
        # Converge 
        if i >= 5 and (cost[-2] - cost[-1]) <= precision:
            break
        
    index = np.arange(1,len(cost)+1,1)
    plt.plot(index,cost)
    
    return w
```

```python
df = pd.read_csv('wdbc.data', header=None)
base_names = ['radius', 'texture', 'perimeter', 'area', 'smooth', 'compact', 'concav', 
                 'conpoints', 'symmetry', 'fracdim']
names = ['m' + name for name in base_names]
names += ['s' + name for name in base_names]
names += ['e' + name for name in base_names]
names = ['id', 'class'] + names
df.columns = names
df['color'] = pd.Series([(0 if x == 'M' else 1) for x in df['class']])
my_color_map = mpl.colors.ListedColormap(['r', 'g'], 'mycolormap')

from sklearn.linear_model import LogisticRegression

c1 = 'mradius'
c2 = 'mtexture'

clf = LogisticRegression()
clf.fit(df[[c1, c2]], df['color'])

plt.scatter(df[c1], df[c2], c = df['color'], cmap=my_color_map)
plt.xlabel(c1)
plt.ylabel(c2)

x = np.linspace(df[c1].min(), df[c1].max(), 1000)
y = np.linspace(df[c2].min(), df[c2].max(), 1000)
xx, yy = np.meshgrid(x,y)
predictions = clf.predict(
        np.hstack(
            (xx.ravel().reshape(-1,1), yy.ravel().reshape(-1,1))
        ))
predictions = predictions.reshape(xx.shape)

plt.contour(xx, yy, predictions, [0.0])
```
