# question 3

# a

function is given below:

```r
haha = function(x){
  dataset=sample(x,42)
  training = dataset[1:23]
  test = dataset[24:42]
  
  result=data.frame(lower_bound=numeric(1),test_mean=numeric(1),upper_bound=numeric(1))

  result$lower_bound[1]=mean(training)-qt(.025, df=length(training)-1)*sd(training)/sqrt(length(training))
  result$test_mean[1]=mean(test)
  result$upper_bound[1]=mean(training)+qt(.025, df=length(training)-1)*sd(training)/sqrt(length(training))

  return(result)
  }

haha(h$spamscore)
```

Specifically, $ Y_i $ is heteroscedastic, because its variance changes as i changes, which violates linear regression model assumption. 

Code used to do experiment: 

```r
each_time=function(){
  
  result=data.frame(X=numeric(100),Y=numeric(100))
  result$X=rnorm(100)
  
  for (i in 1:100){
    result$Y[i] = sum(result$X[1:i])
  }
  
  #return(result)
  
  xline=lm(result$X ~ c(1:100))
  yline=lm(result$Y ~ c(1:100))
  
  plot(result$X,col="red",xlab = "i",ylab="X_i, Y_i",ylim=range(c(-40:40)))
  points(result$Y,col="blue")
  
  abline(xline,col="red",ylim=range(c(-40:40)))
  abline(yline,col="blue",ylim=range(c(-40:40)))
  legend('topright',names(result),lty=1,col=c("red","blue"),bty='n',cex=.5)
  
  return(result)
}

result=each_time()
result=each_time()
result=each_time()
result=each_time()
plot(result$Y)
```
