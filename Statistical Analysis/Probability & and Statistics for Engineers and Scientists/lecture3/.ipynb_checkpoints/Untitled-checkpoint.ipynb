(a) 

binomial random number generator

Definition of parameters: 

n: number of independent trials  

p: probabality of success on each trial.

For example, n = 100, p =0.5 means how many successful trials out of 100, each of which independently has 50% to succeed. 


```r
# function for problem a:
binomial = function(n,p){sum(sample(0:1,n,replace=T,prob=c(1-p,p)));}

# backtest with n=20 and p =0.5, 1000 times repetition:
# rbinom result:
hist(rbinom(size=20,n=1000,p=0.5))

# my result: 
a=vector('numeric')
for(i in 1:1000){a=append(a,binomial(20,0.5));}
hist(a)
```

(b).

gemetric random number generator

Definition of parameters:  

p: probabality of success on each trial.

For example, p =0.5 means how many trials until the first successful, inclusive, each of which independently has 50% to succeed. 

```r
# function for problem b
geometric = function(p){
  sum =0;
  while(TRUE){
    if(sample(0:1,1,replace=T,prob=c(1-p,p))==1){
      return(sum+1);
    }
    
    else{
      sum=sum+1;
    }
  }
}


# backtest with p = 0.1, repetition of 1000 times: 
# NOTICE: run this whole block (In [*]) some times, less than 5, will guarantee that the first bars of 
#         the two histograms close to each other


# rgeom(p=0.1)
hist(rgeom(1000,0.1))

# my version of geometric function:
a=vector('numeric')
for(i in 1:1000){a=append(a,geometric(0.1));}
hist(a)
```

(c).

hypergemetric random number generator

Definition of parameters: 

M1: number of units within the population that are labeled 1

M2: number of units within the population that are labeled 0

N: the sample size taken from M1 + M2.

For example, M1 = 10, M2= 100, N =5 means how many units that are labeled 1 does a sample size of 5 has when taking from a population of 110 units, which contains 10 units that are labeled as 1. 

```r
# function for problem c 

hypergeometric = function(m1,m2,n){
  # shuffle it twice to ensure complete randomness of the sample. 
  shuffle_sample = sample(c(rep(1,m1),rep(0,m2)),m1+m2,replace=F)
  sum(sample(shuffle_sample,n,replace=F))
}

# backtest with m1=50, m2=100, n=5, repetition of 1000 times 

# rhyper(1000,50,100,5)
hist(rhyper(1000,50,100,5))

# my own version of hypergeometric
a=vector('numeric')
for(i in 1:1000){a=append(a,hypergeometric(m1=50,m2=100,n=5));}
hist(a)
```
