# 2.3.5

a) Simulate 10 replications of the experiment that records the number of times a six occurs in two die rolls.

```r
dice <- expand.grid(X1 = 1:6, X2 = 1:6)
y <- (dice$X1 == 6) + (dice$X2 == 6)
sample(y, 10, replace = T)
```

b) Find the relative frequencies of each outcome in the sample space of the experiment that records the number of times a six occurs in two die rolls, based on 10,000 replications.

```r
x <- sample((dice$X1 == 6) + (dice$X2 == 6), 10000, replace = T)
table(x)/10000
```

c) Construct a histogram of the relative frequencies and line graph of the probability mass function.

```r
hist(x, seq(-0.5, 2.5, 1), freq = F)
pr <- table(y)/36
lines(0:2, pr, type = "p", col="red")
lines(0:2, pr, type = "h", col="red")
```

# 2.4.3

We measure the gas mileage of three different cars. Each of the cars can have mileage 50, 51, 52, or 53 mpg with equal probability.
a) Find the probability mass function of the experiment that records the average mileage.

```r
S <- expand.grid(50:53, 50:53, 50:53)
means <- rowMeans(S)
pmf <- table(means)/length(S$Var1)
pmf
```

b) Compute the probability that the average gas mileage is at least 52 mpg.

```r
sum(pmf[as.numeric(names(pmf)) >= 52])
```

# 2.4.11

```r
A <- c(4,4,4,4,0,0)
B <- rep(3,6)
C <- c(2,2,2,2,6,6)
D <- c(5,5,5,1,1,1)
```

Find the probabilities that A > B, B > C, C > D, D > A.

```r
S1 <- expand.grid(A = A, B =B)
mean(S1$A > S1$B)
```

```r
S2 <- expand.grid(B = B, C = C)
mean(S2$B > S2$C)
```

```r
S3 <- expand.grid(C = C, D = D)
mean(S3$C > S3$D)
```

```r
S4 <- expand.grid(D = D, A = A)
mean(S4$D > S4$A)
```
